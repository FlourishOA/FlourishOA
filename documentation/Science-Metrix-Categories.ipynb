```python
#Import file
flourishoa = pd.read_csv('api_journal8-23-17.csv', sep=",", encoding='cp1252', 
                         error_bad_lines=False, warn_bad_lines=True)
```

```python
#Import file
sci_metrix = pd.read_csv('sm-journal-classifications.csv', sep=",")
```

```python
# Merge where issn's match
issn_merge = pd.merge(flourishoa, sci_metrix, on='issn', how='inner')
```

```python
# Merge where flourishoa ISSN matches Sci-M ESSN
essn_merge = pd.merge(flourishoa, sci_metrix, left_on='issn', right_on='essn', how='inner')
```

```python
# Drop unneeded columns
issn_merge = issn_merge.drop(['is_hybrid', 'category', 'url', 'Source_title', 'essn', 'smsid', 'Unnamed: 8'], axis=1)
```

```python
# Drop unneeded columns
essn_merge = essn_merge.drop(['is_hybrid', 'category', 'url', 'Source_title', 'issn_y', 'essn', 'smsid', 'Unnamed: 8'], axis=1)
```

```python
# Rename columns
header = ['issn','journal_name','pub_name','domain','field','subfield','sm_call_no']
issn_merge.columns = header
```

```python
essn_merge.head(1)
```

```python
# Rename columns
essn_merge.columns = header
```

```python
# concatenate the two dataframes
sm_category = pd.concat([issn_merge, essn_merge])
# Reset the index, dropping original
sm_category = sm_category.reset_index(drop=True)
sm_category
```

```python
# save to csv
sm_category.to_csv('scimet_categories.csv', sep=',')
```
